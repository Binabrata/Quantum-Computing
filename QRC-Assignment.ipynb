```python
df = df.drop(numerical_cols, axis = 1)
df = pd.concat([df_numerical_cols, df], axis = 1)
df
```

```python
X = df.drop('Class', axis=1)
y = pd.DataFrame(df['Class'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, shuffle=True, stratify=y)

print(y_train.value_counts())
print(y_test.value_counts())
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.circuit.library import ZZFeatureMap

import numpy as np
```

```python
n_qubits = 20
```

```python
qc = QuantumCircuit(n_qubits)

feature_map = ZZFeatureMap(feature_dimension=n_qubits, entanglement="full", reps=2)
qc.append(feature_map, range(n_qubits))

for i in range(n_qubits):
  qc.rx(np.random.uniform(0, 2 * np.pi), i)
  qc.rz(np.random.uniform(0, 2 * np.pi), i)
  qc.cx(i, (i + 1) % n_qubits)

#qc.measure_all()

qc.draw()
```

```python
print(qc.num_parameters)
```

```python
print(qc.parameters)
```

```python
n_shots = 1024

backend = Aer.get_backend('statevector_simulator')
```

```python
train_outputs = []
for i in range(len(X_train)):
  compiled_circuit = transpile(qc, backend)
  compiled_circuit = compiled_circuit.assign_parameters(X_train.iloc[i])

  result = backend.run(compiled_circuit, shots=n_shots).result()
  statevector = result.get_statevector()

  train_outputs.append(np.abs(statevector)**2)

train_outputs
```

```python
len(train_outputs)
```

```python
len(y_train)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(max_iter=1000)
classifier.fit(train_outputs, y_train)
```

```python
test_outputs = []
for i in range(len(X_test)):
  compiled_circuit = transpile(qc, backend)
  compiled_circuit = compiled_circuit.assign_parameters(X_test.iloc[i])

  result = backend.run(compiled_circuit, shots=n_shots).result()
  statevector = result.get_statevector()
  
  test_outputs.append(np.abs(statevector)**2)

test_outputs
```

```python
y_pred = classifier.predict(test_outputs)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix, roc_curve, auc

print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print(f'Accuracy Score: {accuracy_score(y_test, y_pred)}')
print(f'Precision Score: {precision_score(y_test, y_pred, average="macro")}')
print(f'Recall Score: {recall_score(y_test, y_pred, average="macro")}')
print(f'F1 Score: {f1_score(y_test, y_pred, average="macro")}')
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
print(f'ROC AUC: {roc_auc}')
```
